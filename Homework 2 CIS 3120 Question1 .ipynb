```python
#import libraries
import requests 
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#set the url
url = "https://www.nfl.com/stats/player-stats/"
#Take the resquests
page = requests.get(url)
#status code 200 = request was successful
page.status_code
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
#Find table 
table = soup.find('table')
print(table)
```

```python
table = soup.find('tr')
print(table)
```

```python
#find the table element
table = soup.find('table')  

#find all the table headers
headers = table.find_all('th') 
for header in headers:
    print(header.text)
```

```python
# find all the table headers and store them in a list
headers = []
for th in table.find_all('th'):
    headers.append(th.text)

# remove the unwanted header
headers = headers[0:6]

# create a new DataFrame with the remaining headers
df = pd.DataFrame(columns=headers)
df.head()
```

```python
url = "https://www.nfl.com/stats/player-stats/"
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")
table = soup.find("table")

df = pd.read_html(str(table))[0]
print(df)
```
